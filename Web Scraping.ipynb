```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests
```

```python
url = "https://fastestlaps.com/tracks/adm-miachkovo"

page = requests.get(url)

print(page.status_code)
```

```python
page.content
Soup = BeautifulSoup(page.content,"html.parser")
```

```python
print(Soup.prettify())
```

```python
Body2 = Soup.tbody 
Bodies2 = []  


rows = Body2.find_all("tr")

for row in rows:
    cell_values = []  
    cells = row.find_all("td")  
    for cell in cells:
        cell_text = cell.get_text(strip=True)  
        cell_values.append(cell_text) 
    Bodies2.append(cell_values)  

df = pd.DataFrame(Bodies2, columns = ["Rank", "Vehicle","Driver","Time", "PS"])

df.drop("Rank", axis =1, inplace = True)

df
```

```python
Body2 = Soup.tbody
Body2

Bodies5 = []

for x in Body2.find_all("tr"):
    td_tags = x.find_all("td")
    td_val = [y.text for y in td_tags]
    Bodies5.append(td_val)

df = pd.DataFrame(Bodies5, columns = ["Rank", "Vehicle","Driver","Time", "PS"])

df.drop("Rank", axis =1, inplace = True)
df
```
